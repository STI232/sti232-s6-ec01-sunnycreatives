```python
import pandas as pd
import matplotlib.pyplot as plt
df = pd.read_csv("v2_Latest_Data_Science_Salaries.csv")
df.head(10)
```

### Propuesta de solución


```python
counter = df['Job Title'].value_counts()
print (counter)
```

```python
 et_frecuency = df ['Employment Type'].nunique()
print( et_frecuency )
```

```python
 et_frecuency = df ['Employment Type'].value_counts()
print( et_frecuency )
```

```python
ExpertiseL_types = df ['Expertise Level'].value_counts()
print (ExpertiseL_types)
```

```python
group_expert_expertise = df.groupby('Expertise Level') 
mean_group_expert_expertise = group_expert_expertise ['Salary in USD'].mean()
print (mean_group_expert_expertise)
```

```python
ExperienceL_types = df ['Experience Level'].value_counts()
print (ExperienceL_types)
```

```python
group_expert_experience = df.groupby('Experience Level') 
mean_group_expert_experience = group_expert_experience ['Salary in USD'].mean()
print (mean_group_expert_experience)
```

```python
divisa_filter = df [df['Salary Currency'].notna()]
conversion = {}
for index, row in divisa_filter.iterrows():
    currency = row ['Salary Currency']
    rate = row ['Salary']
    conversion[currency] = rate
for currency, rate in conversion.items():
    print (f'Divisa: {currency}, Tasa de conversión a USD: {rate}')
```

```python
residence = df['Employee Residence']
countries = set (residence)
countries_number = len (countries)
print(f'Hay empleados en{countries_number} países')
```

```python
mean_salary = df.groupby('Job Title')['Salary'].mean()
print (mean_salary)
```

```python
mean_salary_country = df.groupby('Employee Residence')['Salary'].mean()
print (mean_salary_country)
```

```python
max_salary = df['Salary in USD'].max()
max_salary_register = df[df['Salary in USD'] == max_salary]

print(f"Max salario (USD): {max_salary}")
print("Registro correspondiente:")
print(max_salary_register)


min_salary = df['Salary in USD'].min()
min_register_salary = df[df['Salary in USD'] == min_salary]

print(f"\nMin salario (USD): {min_salary}")
print("Registro correspondiente:")
print(min_register_salary)
```

```python
salary_mean_role = df.groupby(['Company Size'])['Salary'].mean().reset_index()

print (salary_mean_role)
```
